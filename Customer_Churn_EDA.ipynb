```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
# Fill missing TotalCharges with MonthlyCharges * tenure where TotalCharges is NaN
df['TotalCharges'] = df['TotalCharges'].fillna(df['MonthlyCharges'] * df['tenure'])
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
df['SeniorCitizen'].value_counts()
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()


# numeric_cols = [c for c in numeric_cols if c not in ['SeniorCitizen']]

print("The Numerical Columns:", numeric_cols)

# ────────────────────────────────────────────────
for col in numeric_cols:


    fig = make_subplots(
        rows=1, cols=2,
        column_widths=[0.75, 0.25],
        subplot_titles=("Histogram + Normal curve", "Boxplot"),
        horizontal_spacing=0.06
    )

    # ─── Histogram + Normal curve ────────────────────────────────
    fig.add_trace(
        go.Histogram(
            x=df[col],
            nbinsx=60,
            marker_color="#5c9bd1",
            opacity=0.75,
            name="Histogram",
            histnorm='probability density'
        ),
        row=1, col=1
    )

    mu = df[col].mean()
    sigma = df[col].std()


    x_curve = np.linspace(df[col].min() - 5, df[col].max() + 5, 300)
    y_curve = stats.norm.pdf(x_curve, mu, sigma)

    fig.add_trace(
        go.Scatter(
            x=x_curve,
            y=y_curve,
            mode='lines',
            line=dict(color='red', width=2.4, dash='dash'),
            name=f'Normal (μ={mu:.2f}, σ={sigma:.2f})'
        ),
        row=1, col=1
    )

    # ─── Boxplot ─────────────────────────────────────────────────
    fig.add_trace(
        go.Box(
            y=df[col],
            marker_color="#ed7d31",
            name="Boxplot",
            boxmean=True,
            boxpoints=False
        ),
        row=1, col=2
    )

    #──────────────────────────────────────────
    fig.update_layout(
        title_text=f"{col} — Histogram + Normal Distribution & Boxplot",
        title_x=0.5,
        height=520,
        width=1400,
        showlegend=True,
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
        template="plotly_white",
        hovermode="closest",
        bargap=0.07,
        font=dict(size=13)
    )

    fig.update_xaxes(title_text=col, row=1, col=1)
    fig.update_xaxes(title_text=col, row=1, col=2)
    fig.update_yaxes(title_text="Density", row=1, col=1)
    fig.update_yaxes(title_text="", row=1, col=2)

    fig.show()
```

```python
object_cols = df.select_dtypes(include=['object']).columns.tolist()
object_cols = [c for c in object_cols if c.lower() != 'customerid']

print("Categorical columns:", object_cols)
for col in object_cols:

    n_categories = df[col].nunique()

    fig = px.histogram(
        df,
        x=col,
        color="Churn",
        barmode="group",
        title=f"{col} — Count by Churn",
        template="plotly_white",
        color_discrete_sequence=["#636EFA", "#EF553B"],  # أزرق لـ No، أحمر لـ Yes
        text_auto=True,
        category_orders={col: df[col].value_counts().index.tolist()}
    )

    fig.update_layout(
        xaxis_title=col,
        yaxis_title="Number of Customers",
        bargap=0.18,
        bargroupgap=0.12,
        legend_title_text="Churn",
        height=500 + (n_categories * 8),
        width=950 if n_categories <= 10 else 1200,
        font=dict(size=13),
        title_x=0.5,
        hovermode="x unified"
    )

    fig.update_xaxes(
        tickangle= -45 if n_categories > 6 else 0,
        tickfont=dict(size=11)
    )

    fig.show()
```
